**Step 1: Import Libraries**

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import (
    ModelCheckpoint,
    EarlyStopping,
    ReduceLROnPlateau
)

print(tf.__version__)
```

**Step 3: Define Dataset Paths**

```python
train_dir = "/kaggle/input/datasets/ascanipek/eyepacs-aptos-messidor-diabetic-retinopathy/dr_unified_v2/dr_unified_v2/train"

val_dir = "/kaggle/input/datasets/ascanipek/eyepacs-aptos-messidor-diabetic-retinopathy/dr_unified_v2/dr_unified_v2/val"

test_dir = "/kaggle/input/datasets/ascanipek/eyepacs-aptos-messidor-diabetic-retinopathy/dr_unified_v2/dr_unified_v2/test"
```

**Step 4: Check Dataset Structure**

```python
for folder in [train_dir, val_dir, test_dir]:

    print("\n", folder)

    total = 0

    for c in sorted(os.listdir(folder)):

        n = len(os.listdir(os.path.join(folder,c)))

        print(f"Class {c}: {n}")

        total += n

    print("Total:", total)
```

**Step 5: Image preprocessing and augmentation**

```python
IMG_SIZE = 224
BATCH_SIZE = 32
NUM_CLASSES = 5
```

Training generator:

```python
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)
```

Validation/Test:

```python
test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)
```

**Step 6: Create Generators**

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

val_generator = test_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
```

**Step 7: Display Class Labels**

```python
print(train_generator.class_indices)
```

**Step 8: Build MobileNetV2 Model**

```python
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)
```

Freeze layers:

```python
base_model.trainable = False
```

**Add classification layers:**

```python
x = base_model.output

x = GlobalAveragePooling2D()(x)

x = Dense(512,activation='relu')(x)

x = Dropout(0.5)(x)

output = Dense(
    NUM_CLASSES,
    activation='softmax'
)(x)

model = Model(
    inputs=base_model.input,
    outputs=output
)
```

**Step 9: Compile Model**

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

**Step 11: Callbacks**

```python
checkpoint = ModelCheckpoint(
    "mobilenetv2_best.keras",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max",
    verbose=1
)

early_stop = EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.2,
    patience=2,
    verbose=1
)
```

**Step 12: Train Model**

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=3,
    callbacks=[
        checkpoint,
        early_stop,
        reduce_lr
    ]
)
```

Evaluate on Test Set

```python
results = model.evaluate(test_generator)

print("Test Accuracy:", results[1]*100)
print("Test Loss:", results[0])
```
